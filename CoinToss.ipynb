```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
first = [0]*100000
randomnu = [0]*100000
least = [0]*100000
```

```python
for i in range(100000):
    randcoin = random.randint(0,1000)
    leastheads = 10
    heads = 0
    for k in range(10):
        if(random.randint(0,1)==1):
            heads+=1
    first[i]=heads/10.0
    for j in range(1,1000):
        heads = 0
        for k in range(10):
            if(random.randint(0,1)==1):
                heads+=1
        if(heads<leastheads):
            leastheads=heads
        if(j==randcoin):
            randomnu[i]=heads/10.0
    least[i]=leastheads/10.0
```

```python
plt.hist(first,bins='auto')
plt.title("First Coin Sample Means")
plt.show
```

```python
plt.hist(randomnu,bins='auto')
plt.title("Random Coin Sample Means")
plt.show
```

```python
plt.hist(least,bins='auto')
plt.title("Least Coin Sample Means")
plt.show
```
